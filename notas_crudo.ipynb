O en términos de las condiciones iniciales de las dos acciones fundamentales $R_{_+}$ y $R_{_-}$:

Teniendo en cuenta:

$$\sqrt{R_{_+}R_{_-}}=\sqrt{\frac{(x_o+y_o)}{2}\frac{(x_o-y_o)}{2}}$$

$$\sqrt{R_{_+}R_{_-}}=\sqrt{\frac{x_o^2-y_o^2}{4}}$$

$$2\sqrt{R_{_+}R_{_-}}=\sqrt{x_o^2-y_o^2}$$

Concluimos en:

$$F_{_1}=2\sqrt{R_{_+}R_{_-}}$$

## Y el perímetro?

### Perímetro pensando en ondas

Si tienes la suma de dos ondas $$R_{_+}e^{it}+R_{_-}e^{-it}$$ la velocidad en cualquier punto dado es $$iR_{_+} e^{it}-iR_{_-}e^{-it}$$ Para encontrar la magnitud de la velocidad, multiplicas por el conjugado y tomas la raíz cuadrada, por lo que obtienes $$\sqrt{(iR_{_+} e^{it}-iR_{_-}e^{-it})(-i\bar{R_{_+}} e^{-it}+i\bar{R_{_-}}e^{it})} = \sqrt{R_{_+}\bar{R_{_+}}+R_{_-}\bar{R_{_-}}-R_{_+}\bar{R_{_-}}e^{2it}-\bar{R_{_+}}R_{_-}e^{-2it}}$$ Denotando $\bar{z}$ al conjugado de $z$.

# Una introducción amigable a los números complejos en el trabajo: Tiempo, compás y números.

La idea es hacer una reintroducción amigable (asumiendo que ya conoces lo básico) sobre la idea de los números complejos, usando un ejemplo lúdico y la analogía de una compás en lugar de utilizar la palabra "complejo" cada vez.

Imaginemos una gran compás que utilizamos para colocar símbolos en un plano 2D, donde definimos $r$ y $\theta$.

Para simplificar la notación, llamemos a un "número generador de fase" como el número "compás" simple $e^i$ y llamémoslo $\Theta$.

La compás puede ajustarse a un ángulo entre $-\pi$ y $\pi$, al que llamaremos $\theta$, y se puede usar como argumento en una función (eigen) que denotamos como $\Theta^{\theta}$.

Esto establece el ángulo de destino de nuestra compás (imaginemos que en reposo el lápiz está alineado con el eje horizontal, y desde allí hacemos un desplazamiento de $\theta$ radianes en sentido contrario a las agujas del reloj).

$$\Theta = e^i = \cos(1) + i\sin(1)$$

$$\Theta^{\theta} = e^{i \theta} = \cos(\theta) + i\sin(\theta)$$

## El radio
También introduciremos la idea de longitud desde nuestro centro de coordenadas (donde se coloca el centro de nuestra compás).

Llamaremos a esto $r$, el radio del círculo que podemos trazar para ajustar la compás a esta longitud:

$$ r \in \mathbb{R} $$

O más generalmente, $r$ también está en $\mathbb{C}$, lo que significa que puedo usarlo para cambiar la fase de $\Theta^{\theta}$.

Por razones históricas, relacionadas con cómo se descubrió/inventó este tipo de número, esto se llama formalmente un Número Complejo.

$$ r\Theta^{\theta} \in \mathbb{C} $$

El $i$ usado como exponente aquí, para simplificar la notación, indica que el argumento de la función $exp()$ es $i\theta=\sqrt{-1}\theta$.

$$\theta_k=\omega_k k$$

## El juego del calendario circular

Ahora, sígueme...

* Usemos $\omega_k=\frac{2\pi}{T_k}$ como una medida de cuántos símbolos $T_k$ queremos colocar distribuidos en un arco de $2\pi$ radianes para representar como un conjunto de números compás la secuencia de enteros denotada como $k$. (mencionemos la relación introductoria raíces de la unidad, aritmética modular y periodicidad).

* El uso de $\omega$ aquí puede parecer extraño, porque no nos referimos exactamente a la velocidad angular, pero ten en cuenta que esto es solo un truco de notación que tendrá más sentido más adelante.

* Para este experimento, consideremos, por simplicidad, el ejemplo de datos horarios de un conjunto de condiciones climáticas aleatorias a lo largo de muchos años.

Dado que estamos trabajando con tiempo discreto, siempre tenemos un índice principal $n$ que llamaremos $h \in \mathbb{N}_0$ para este ejemplo.

Supongamos que $h$ es la hora, $d$ es el día, $m$ es el mes y $y$ es el año. Establecemos una condición inicial para este mapa.

* La cantidad total de años que vamos a analizar (7 en este ejemplo, desde 2023 hasta 2029), por lo que imaginamos un reloj módulo 7 que va de 0 a 6, donde deseamos colocar los datos por hora para cada año.

$$T_y=N$$

$$\omega_y=\frac{2\pi}{T_y}$$

* El número de meses en un año

$$T_m=12$$

$$\omega_m=\frac{2\pi}{T_m}$$

* Usaremos $T_d=31$ para el número total de días en un mes determinado, ya que el número máximo de días en un mes dado estará entre 28 y 31. Esto puede obtenerse a partir de una función que toma el año y el mes y devuelve el número total de días en ese mes. Elegir 31 como un $T_d$ fijo facilita ver los meses que tienen 30 o 28/29 días porque algunos espacios del reloj módulo 31 no tendrán ninguna hora alrededor de ellos.

$$T_d=monthrange(2023, 12)=31$$

* o para un mes 2 en un año bisiesto

$$T_d=monthrange(2024, 2)=29$$

$$\omega_d=\frac{2\pi}{T_d}$$

* El número total de horas en un día

$$T_h=24$$

$$\omega_h=\frac{2\pi}{T_h}$$

* El número total de minutos en una hora

$$T_{min}=60$$

$$\omega_{min}=\frac{2\pi}{T_{min}}$$

* El número total de segundos en un minuto

$$T_{sec}=60$$

$$\omega_{sec}=\frac{2\pi}{T_{sec}}$$

* Etc. Podríamos seguir hasta los femtosegundos o la escala de tiempo de Planck, pero no importa por ahora.

## $\hat A$: Obtenemos el atlas sumando los números compás

Aprovechamos que sumar números complejos es tan fácil como sumar vectores, así que concatenamos los diferentes engranajes de nuestros relojes usando los diferentes radios $r$ en nuestra representación.

$$\hat A = \hat y + \hat m + \hat d + \hat h$$

### Desglosando el atlas para obtener una expresión que parece más complicada nos ayuda a recordar que solo estamos sumando vectores en la punta de nuestros conjuntos de compáss de haz finito:

Podemos desglosar esto a su núcleo, pero ten en cuenta que solo estamos sumando vectores, lo que significa dónde colocar el lápiz pensando "ortogonalmente" al eje horizontal y vertical. Formalmente, estas proyecciones del vector en los ejes se llaman partes "real" e "imaginaria" del número complejo, pero de nuevo, estamos buscando expresiones más amigables...

$$\hat A = r_y e^{i\frac{2\pi}{T_y}y } + r_m e^{i\frac{2\pi}{T_m}m } + r_d e^{i\frac{2\pi}{T_d}d} + r_h e^{i\frac{2\pi}{T_h}h }$$

$$\hat{A} = r_y \left( \cos \left( \frac{2\pi}{T_y} y \right) + i \sin \left( \frac{2\pi}{T_y} y \right) \right) + r_m \left( \cos \left( \frac{2\pi}{T_m} m \right) + i \sin \left( \frac{2\pi}{T_m} m \right) \right) + r_d \left( \cos \left( \frac{2\pi}{T_d} d \right) + i \sin \left( \frac{2\pi}{T_d} d \right) \right) + r_h \left( \cos \left( \frac{2\pi}{T_h} h \right) + i \sin \left( \frac{2\pi}{T_h} h \right) \right) $$

# Un ejemplo de expresión más general $(x_t,y_t,z_t)$ como cualquier secuencia 3D sobre la base de tiempo discreta $t$

$$\hat A_t = \hat x + \hat y + \hat z + \hat t $$

$$\hat A_t = r_x e^{i\frac{2\pi}{T_x}x } + r_y e^{i\frac{2\pi}{T_y}y } + r_z e^{i\frac{2\pi}{T_z}z} + r_t e^{i\frac{2\pi}{T_t}t }$$

# Lo que la función $exp()$ está haciendo realmente es una historia importante.

* Las funciones exponenciales pueden representarse mediante su serie de Maclaurin (en la práctica, utilizamos una cantidad finita de términos y trucos para mejorar la precisión):

$$\exp(\theta) = e^{\theta}=\sum_{n=0}^{\infty} \frac{\theta^n}{n!} = 1 + \frac{\theta}{1!} + \frac{\theta^2}{2!} + \frac{\theta^3}{3!} + \cdots$$

$$\Theta^{\theta}=e^{i\theta} = \cos(\theta) + i\sin(\theta)$$

* Si desglosamos esta suma, veremos que los exponentes pares se asocian con números reales y los exponentes impares tendrán la $i$ adjunta a ellos.

Esto significa que vamos a tener rotaciones sucesivas de $90º$ mientras:

Paso 1 (n = 0):
$$ \frac{{(i\theta)^0}}{{0!}} = 1 $$

Paso 2 (n = 1):
$$ \frac{{(i\theta)^1}}{{1!}} = i\theta $$

Paso 3 (n = 2):
$$ \frac{{(i\theta)^2}}{{2!}} = -\frac{{\theta^2}}{{2}} $$

Paso 4 (n = 3):
$$ \frac{{(i\theta)^3}}{{3!}} = -\frac{{i\theta^3}}{{6}} $$

Paso 5 (n = 4):
$$ \frac{{(i\theta)^4}}{{4!}} = \frac{{\theta^4}}{{24}} $$

Combinando todos los términos, obtenemos:

$$ \exp(i\theta) = 1 + i\theta - \frac{{\theta^2}}{{2}} - \frac{{i\theta^3}}{{6}} + \frac{{\theta^4}}{{24}} - \ldots $$

$$e^{i \theta} = \sum_{n=0}^{\infty} \frac{(i \theta)^n}{n!} $$

* Podemos recordar que el coseno es el término par y el seno el término impar al ver que:

$$\cos(\theta) = \sum_{n=0}^{\infty} (-1)^n \frac{\theta^{2n}}{(2n)!} = 1 - \frac{\theta^2}{2!} + \frac{\theta^4}{4!} - \frac{\theta^6}{6!} + \cdots$$

$$\sin(\theta) = \sum_{n=0}^{\infty} (-1)^n \frac{\theta^{2n+1}}{(2n+1)!} = \theta - \frac{\theta^3}{3!} + \frac{\theta^5}{5!} - \frac{\theta^7}{7!} + \cdots$$

* Entonces, un desglose completo de lo que estamos haciendo comienza con entender la dinámica de esta suma:

$$e^{i\theta} = \sum_{n=0}^{\infty} \left[ \left(-1\right)^n \frac{\theta^{2n}}{(2n)!} \right] + i \sum_{n=0}^{\infty} \left[ \left(-1\right)^n \frac{\theta^{2n+1}}{(2n+1)!} \right]$$
